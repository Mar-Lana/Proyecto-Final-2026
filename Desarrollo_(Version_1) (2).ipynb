### **1). INTRODUCCIÓN:**



**- Breve Estado del Arte:** Desde la perspectiva de las Ciencias de la Computación, el análisis de grandes volúmenes de datos ha evolucionado de modelos estadísticos descriptivos hacia el Aprendizaje Automático (en inglés “Machine Learning”). En la última década, el auge de las arquitecturas basadas en conjuntos de árboles de decisión, como Random Forest y XGBoost, ha demostrado una precisión superior en la predicción de eventos multivariables. Investigaciones internacionales han validado que estos algoritmos permiten procesar eficazmente datos heterogéneos (fechas, coordenadas geográficas y condiciones climáticas) [1], [2]. Asimismo, el movimiento de Gobierno Abierto (en inglés “Open Government”) ha estandarizado el uso de infraestructuras de Datos Abiertos [3], permitiendo que herramientas de procesamiento en la nube, como Google Colab y bibliotecas de Python (Pandas y Scikit-learn) [4],[5], sean utilizadas para transformar registros públicos en modelos predictivos de alta disponibilidad y bajo costo de implementación.

En el ámbito de la Seguridad Vial, el paradigma dominante es la estrategia "Vision Zero", originada en Suecia, la cual sostiene que la siniestralidad es una falla del sistema y no una consecuencia inevitable del error humano. Este enfoque ha impulsado a ciudades modernas a crear sistemas de monitoreo proactivos [6]. En Argentina, si bien se han consolidado Observatorios Viales a nivel nacional y provincial para la elaboración de estadísticas anuales [7], la aplicación de técnicas de Ciencia de Datos sobre datasets municipales sigue siendo un área poco explorada. La disponibilidad de datos abiertos en la Ciudad de Corrientes desde 2018 [8] presenta una oportunidad única para cerrar la brecha entre la recolección de datos y la implementación de políticas públicas preventivas basadas en evidencia científica.

### 1.2. Random Forest y XGBoost

Dentro de las arquitecturas basadas en conjuntos de árboles de decisión, se destacan dos algoritmos por su capacidad de generalización:

*   **Random Forest:** Es un algoritmo de aprendizaje supervisado que utiliza una combinación de árboles de decisión para mejorar la precisión y controlar el sobreajuste (overfitting).
*   **XGBoost (eXtreme Gradient Boosting):** Se fundamenta en un marco de optimización de gradiente, demostrando una eficiencia superior en el procesamiento de datos a gran escala y la captura de relaciones no lineales entre variables ambientales y viales.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
import pandas as pd
import numpy as np


# =================================================================
# 3. INFORME DE VERIFICACIÓN DE LA CALIDAD (Data Quality)
# =================================================================

print(f"\n--- INFORME DE CALIDAD DE DATOS ---")

# A. Detección de Valores Faltantes
nulos = df[temporales + espaciales + ambiental + infraestructura + dimension + gravedad].isnull().sum()
print(f"Valores nulos críticos:\n{nulos[nulos > 0]}")

# B. Consistencia Geográfica (Corrientes Capital)

def limpiar_coordenada_final(valor):
    """
    Limpia, normaliza y escala coordenadas geográficas para la Ciudad de Corrientes,
    marcando valores faltantes (como 'S/D') y coordenadas en cero ('0' o '0.0')
    como NaN para su posterior geocodificación.
    """
    # 1. Filtro temprano: Elimina nulos, etiquetas vacías, 'S/D' y cualquier variante de cero (0, 0.0)
    if pd.isna(valor):
        return np.nan

    valor_str = str(valor).strip().upper()
    # AGREGAMOS -40, -61, 40 y 61 a los filtros tempranos de descarte
    if valor_str in ['S/D', '0', '0.0', 'S/N', 'NAN', '', '-40', '-40.0', '-61', '-61.0', '40', '61']:
        return np.nan

    try:
        # Reemplaza comas por puntos para consistencia decimal
        s = valor_str.replace(',', '.')

        # Corrección de formato con múltiples puntos (ej. -2.748.228 -> -2.748228)
        partes = s.split('.')
        if len(partes) > 2:
            s = partes[0] + partes[1] + '.' + "".join(partes[2:])

        num = float(s)

        # 2. Filtro tardío: Por si la conversión numérica dio exactamente cero
        if num == 0.0:
            return np.nan

        # Filtro tardío por si el float da el valor dummy de Buenos Aires
        if num == 0.0 or num == -40.0 or num == -61.0:
            return np.nan

        # Si la latitud quedó guardada como -2748.xxxx, la dividimos por 100 para llevarla a -27.48xxxx
        if num <= -100.0 or num >= 100.0:
            num = num / 100.0

        # Ajuste secundario si el número quedó truncado en un solo dígito (ej: -2.748 -> -27.48)
        elif -4.0 < num < 0.0:
            num = num * 10.0

        return num
    except Exception:
        # Retorna NaN ante cualquier error de conversión o formato de texto inválido
        return np.nan


# Filtramos coordenadas que no pertenezcan a la región aproximada de la ciudad
df['latitud'] = df['latitud'].apply(limpiar_coordenada_final)
df['longitud'] = df['longitud'].apply(limpiar_coordenada_final)
# Ahora filtramos los que quedaron fuera del rango de la Ciudad de Corrientes
fuera_rango = df[df['latitud'].notna() & df['longitud'].notna() &
                 ((df['latitud'] > -27.3) | (df['latitud'] < -27.6) |
                  (df['longitud'] > -58.6) | (df['longitud'] < -58.9))]

print(f"Registros con coordenadas fuera de Corrientes (posibles errores): {len(fuera_rango)}")
display(fuera_rango[['id_feu', 'nombre_via', 'altura_km', 'entre_calle_1', 'entre_calle_2', 'latitud', 'longitud']])




# =================================================================
# 4. INFORME DE EXPLORACIÓN DE LOS DATOS (EDA INICIAL)
# =================================================================

# Preparación para visualización
df['siniestro_fecha'] = pd.to_datetime(df['siniestro_fecha'], errors='coerce')
df['mes'] = df['siniestro_fecha'].dt.month
df['dia_semana'] = df['siniestro_fecha'].dt.day_name()

# A. Análisis Estadístico: Frecuencia Mensual
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='mes', hue='mes', palette='viridis', legend=False) # Applied the suggested fix
plt.title('Distribución Mensual de Siniestros en Corrientes (2018-2025)')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Accidentes')
plt.show()

# B.1 Visualización de Patrones Temporales: Día de la Semana
orden_dias = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='dia_semana', order=orden_dias, hue='dia_semana', palette='magma', legend=False) # Applied the suggested fix
plt.title('Siniestralidad por Día de la Semana')
plt.show()

# B.2 Visualización de Patrones Temporales: Análisis por Franja Horaria
df['hora_neta'] = pd.to_numeric(df['siniestro_hora'].astype(str).str.split(':').str[0], errors='coerce').astype('Int64')

plt.figure(figsize=(12, 6))
sns.histplot(data=df.dropna(subset=['hora_neta']), x='hora_neta', bins=24, kde=True, color='teal')
plt.title('Distribución de Siniestros por Hora del Día (2018-2025)')
plt.xlabel('Hora del día (0-23)')
plt.ylabel('Cantidad de Siniestros')
plt.xticks(range(0, 24))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# B.3 Visualización de Patrones Temporales: Mapa de Calor Temporal (Día vs. Hora)
# Se crea una tabla pivote para el mapa de calor
# 0=Lunes, 6=Domingo
df['dia_num'] = df['siniestro_fecha'].dt.dayofweek
pivot_table = df.pivot_table(index='dia_num', columns='hora_neta', values='id_feu', aggfunc='count').fillna(0)

# Se traduce índices para el gráfico
dias_es = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

plt.figure(figsize=(15, 6))
sns.heatmap(pivot_table, annot=False, cmap='YlOrRd', yticklabels=dias_es)
plt.title('Densidad de Siniestros: Día de la Semana vs. Hora')
plt.xlabel('Hora del Día')
plt.ylabel('Día de la Semana')
plt.show()

# B.4 Visualización de Patrones Temporales: Evolución Anual (Tendencia Histórica)
# Se extrae el año de la fecha
df['anio'] = df['siniestro_fecha'].dt.year

plt.figure(figsize=(10, 5))
# Filtramos años que podrían estar incompletos (como el actual 2026 si existiera)
df_anual = df[df['anio'].between(2018, 2025)]
sns.lineplot(data=df_anual.groupby('anio').size().reset_index(name='conteo'), x='anio', y='conteo', marker='o', color='red')
plt.title('Evolución Anual de Siniestros (2018-2025)')
plt.xlabel('Año')
plt.ylabel('Total de Siniestros Registrados')
plt.grid(True, alpha=0.3)
plt.show()

# C. Visualización Preliminar: Mapa de Calor (Hotspots)
# Solo usamos registros con coordenadas válidas (únicamente con las latitudes y longitudes completas en el dataset original de la Municipaliad)

# 1. Limpieza específica para el mapa (asegurar que sean números válidos)
# Aplicamos la limpieza a una copia para no alterar el dataframe original
df_mapa = df.copy()
df_mapa['lat_clean'] = df_mapa['latitud'].apply(limpiar_coordenada_final)
df_mapa['lon_clean'] = df_mapa['longitud'].apply(limpiar_coordenada_final)


# 2. Filtrar solo los registros que tienen coordenadas válidas dentro de Corrientes
# Rango aproximado: Lat (-27.4 a -27.6) | Lon (-58.7 a -58.9)
df_geo = df_mapa.dropna(subset=['lat_clean', 'lon_clean'])
df_geo = df_geo[(df_geo['lat_clean'] < -27.3) & (df_geo['lat_clean'] > -27.6) &
                (df_geo['lon_clean'] < -58.6) & (df_geo['lon_clean'] > -58.9)]

# 3. Crear el mapa base centrado en la Ciudad de Corrientes
mapa_corrientes = folium.Map(location=[-27.4692, -58.8306], zoom_start=13, tiles='OpenStreetMap')

# 4. Generar la capa de calor
# Usamos latitud, longitud y podemos usar 'cantidad_de_involucrados' como peso opcional
data_calor = df_geo[['lat_clean', 'lon_clean']].values.tolist()

HeatMap(data_calor,
        radius=15,
        blur=10,
        max_zoom=1,
        gradient={0.4: 'blue', 0.65: 'lime', 1: 'red'}).add_to(mapa_corrientes)

# Mostrar el mapa
mapa_corrientes

# Revisión de Valores Faltantes
nulos = df[temporales + espaciales + ambiental + infraestructura + dimension + gravedad].isnull().sum()
print(f"Valores nulos críticos:\n{nulos[nulos > 0]}")
```

# Reconstrucción Dinámica de Variables Temporales

### 1. Diagnóstico de la Problemática (El Escenario Inicial)

Durante la etapa de Verificación de la Calidad de los Datos para el lote correspondiente al año 2024 (archivo feu-siniestros-2024.csv), se identificó una pérdida crítica de integridad en tres registros específicos (IDs de siniestro: 217505, 217515 y 233620).

Al aplicar el parseador estándar de Pandas (pd.to_datetime) con el parámetro de coerción de errores (errors='coerce'), las fechas de estos tres eventos eran transformadas automáticamente a valores nulos de tipo NaT (Not a Time).

Este comportamiento destructivo en memoria se debió a dos inconsistencias estructurales en la carga original de la base de datos municipal:

1.   *Representación Narrativa de Atributos:* En la columna siniestro_fecha, las fechas de estos eventos estaban registradas en formato de texto natural en español (por ejemplo, el registro del ID 233620 contenía la cadena "9 de julio"), impidiendo su interpretación cronológica directa.
2.   *Degradación de Tipos Numéricos en la Exportación (Efecto Float):* Al abrir la base de datos consolidada, las columnas auxiliares dia y ano sufrieron una distorsión de precisión de coma flotante. El año 2024 fue erróneamente interpretado como 2.024 en dos de los registros, mientras que el día 9 y el 27 se cargaron como flotantes con decimales (9.0 y 27.0), lo cual rompía cualquier algoritmo de concatenación de texto simple.

### 2. Estrategia de Resolución Implementada (El Algoritmo Dinámico)

Para evitar la exclusión metodológica de estos registros (lo cual introduciría un sesgo en el posterior entrenamiento del modelo predictivo de Machine Learning), se diseñó y ejecutó un Algoritmo de Recuperación Física y Reconstrucción Dinámica de Fechas.

El flujo de procesamiento programado se compone de tres fases automatizadas:

**Paso 1: Recuperación Física**
*   Accede directamente a los archivos CSV locales en disco.
*   Recupera la cadena de texto original (ej. "9 de julio") antes de que sea destruida por 'coerce'.

**Paso 2: Limpieza y Normalización de Componentes**
*   Convierte números flotantes corruptos de años (como "2.024") y días (como "27.0") a strings enteros limpios.

**Paso 3: Mapeo Semántico en Español**
*   Utiliza Expresiones Regulares para buscar meses narrativos (como "JULIO").
*   Traduce dinámicamente el mes al índice cronológico estándar ("07").

**Paso 4: Ensamblado y Unificación de Tipo de Dato**
*   Reconstruye la fecha como "09/07/2024" y la convierte al tipo estricto datetime64[ns].
*   Regenera las variables derivadas: mes, día de la semana y es_fin_de_semana.

****

1.  **Recuperación Física Directa (Resguardo de Memoria):** Debido a que la memoria activa del cuaderno de Python sobrescribe los valores corruptos con NaT, el pipeline localiza dinámicamente los archivos CSV originales de almacenamiento físico en el disco (drive/MyDrive/...), extrayendo el texto original de la fecha de forma intacta.
2.  **Normalización Numérica de Coma Flotante (Float Cleansing):** Mediante expresiones regulares, se eliminaron los decimales redundantes de los días y se corrigieron de manera inteligente las representaciones de miles para los años, traduciendo cadenas como "2.024" y "2024.000" a la cadena unificada "2024".
3.  **Mapeo Semántico y Análisis Expresivo:** Se diseñó una función de traducción que analiza semánticamente las palabras de los meses en español utilizando expresiones regulares (capturando patrones del tipo \d+ de [mes]), vinculando cadenas como "enero" o "julio" con sus correspondientes valores numéricos de calendario ("01" y "07").
4.  **Fusión Cronológica de Componentes:** Finalmente, el algoritmo ensambla de manera sintáctica los componentes limpios bajo el estándar de fecha y hora internacional y los re-inyecta al dataframe global como un tipo de dato unificado datetime64[ns].

### 3. Impacto en el Modelo Predictivo

Al unificar las fechas, se recalculan automáticamente de forma perfecta las variables predictoras clave que utilizaremos para alimentar a Random Forest y XGBoost:

*   dia_semana (mapea el comportamiento laboral vs. el fin de semana).
*   es_fin_de_semana (indicador binario de alta relevancia en siniestralidad nocturna).

```python
import pandas as pd
import numpy as np
import re
import os

# =================================================================
# A. ALGORITMO DE RECONSTRUCCIÓN Y PARSEO DINÁMICO
# =================================================================

def reconstruir_fecha_dinamica(row, df_original_csvs):
    """
    Algoritmo de parseo dinámico que corrige anomalías de formato,
    recupera textos originales si la memoria fue corrompida por 'coerce'
    y traduce fechas narrativas en español (ej. '9 de julio').
    """
    meses_espanol = {
        'ENERO': '01', 'FEBRERO': '02', 'MARZO': '03', 'ABRIL': '04', 'MAYO': '05', 'JUNIO': '06',
        'JULIO': '07', 'AGOSTO': '08', 'SETIEMBRE': '09', 'SEPTIEMBRE': '09', 'OCTUBRE': '10',
        'NOVIEMBRE': '11', 'DICIEMBRE': '12'
    }

    id_actual = row.get('id_feu')

    # Intentar obtener la fecha en formato string.
    # Si ya fue convertida a NaT, la buscamos en nuestro diccionario de respaldo recuperado del CSV
    fecha_original = ""
    if id_actual in df_original_csvs and pd.notna(df_original_csvs[id_actual]):
        fecha_original = str(df_original_csvs[id_actual]).strip()
    else:
        fecha_original = str(row.get('siniestro_fecha_original', row.get('siniestro_fecha', '')))

    # Limpieza robusta del año (ej: 2.024 o 2024.000 -> '2024')
    def limpiar_anio(a):
        if pd.isna(a):
            return "2024"
        a_str = str(a).strip()
        # Eliminar decimales de float (ej: 2024.0 -> 2024)
        a_str = re.sub(r'\.0+$', '', a_str)
        # Quitar puntos si es una interpretación de miles errónea (ej: 2.024 -> 2024)
        a_str = a_str.replace('.', '')
        if len(a_str) == 2:
            return f"20{a_str}"
        elif len(a_str) > 4:
            return a_str[:4]
        return a_str

    # Limpieza del día (ej: 27.0 -> '27')
    def limpiar_dia(d):
        if pd.isna(d):
            return None
        return str(d).split('.')[0].strip().zfill(2)

    dia_limpio = limpiar_dia(row.get('dia'))
    anio_limpio = limpiar_anio(row.get('ano'))

    # Detección del mes en la cadena de texto original
    mes_detectado = None
    for mes_nombre, mes_num in meses_espanol.items():
        if mes_nombre in fecha_original.upper():
            mes_detectado = mes_num
            break

    # Si encontramos el día, el año y el mes de forma dinámica, ensamblamos
    if dia_limpio and anio_limpio and mes_detectado:
        fecha_reconstruida = f"{dia_limpio}/{mes_detectado}/{anio_limpio}"
        try:
            return pd.to_datetime(fecha_reconstruida, format='%d/%m/%Y', errors='coerce')
        except:
            pass

    # Si no requiere procesamiento complejo, intentar parseador estándar
    try:
        return pd.to_datetime(row['siniestro_fecha'], errors='coerce')
    except:
        return pd.NaT


# =================================================================
# B. PIPELINE DE COMPARACIÓN Y ORQUESTACIÓN
# =================================================================

if __name__ == "__main__":
    if 'df' in globals():
        print("="*75)
        print("🔍 DETECCIÓN DE FECHAS ERRÓNEAS (ANTES DE LA CORRECCIÓN)")
        print("="*75)

        # Aislamos los 3 registros específicos que fallan
        ids_fallas = [217505, 217515, 233620]
        df_antes = df[df['id_feu'].isin(ids_fallas)].copy()

        # Guardamos el estado "Antes" tal y como está en tu DataFrame actual (donde siniestro_fecha es NaT)
        print("Registros que están fallando (Valores en memoria activa):")
        columnas_antes = ['id_feu', 'siniestro_fecha', 'dia', 'mes', 'ano', 'archivo_origen']
        cols_antes_validas = [c for c in columnas_antes if c in df_antes.columns]
        print(df_antes[cols_antes_validas].to_string(index=False))
        print("-" * 75)

        print("\n" + "="*75)
        print("⚙️ RECUPERANDO TEXTOS ORIGINALES DESDE LOS ARCHIVOS CSV...")
        print("="*75)

        # Diccionario para almacenar los strings originales directo del disco
        textos_originales_csv = {}

        # Agrupamos por archivo de origen para abrirlos solo una vez (más eficiente)
        archivos_afectados = df_antes['archivo_origen'].unique()
        for arch in archivos_afectados:
            # Rutas posibles donde Colab / Drive puede tener el archivo
            rutas_posibles = [
                str(arch),
                f"drive/MyDrive/Proyecto Final de Carrera (2026)/Archivos Siniestros 2018-2025/{arch}"
            ]

            archivo_encontrado = False
            for ruta in rutas_posibles:
                if os.path.exists(ruta):
                    try:
                        # Cargamos temporalmente solo el ID y el String de fecha original del archivo físico
                        df_raw = pd.read_csv(ruta, usecols=lambda x: x.lower().strip() in ['id_feu', 'siniestro_fecha'], encoding='latin1', low_memory=False)
                        df_raw.columns = df_raw.columns.str.lower().str.strip()

                        # Extraemos los strings para nuestros IDs con problemas
                        for id_f in ids_fallas:
                            fila = df_raw[df_raw['id_feu'] == id_f]
                            if not fila.empty:
                                val_original = fila['siniestro_fecha'].values[0]
                                textos_originales_csv[id_f] = val_original
                        archivo_encontrado = True
                        print(f"-> Archivo '{arch}' localizado con éxito. Datos recuperados físicamente.")
                        break
                    except Exception as e:
                        pass

            if not archivo_encontrado:
                print(f"-> [Atención] No se pudo leer el archivo físico '{arch}'. Se usará el fallback del simulador.")

        # Mostramos qué strings físicos de fecha logramos rescatar
        print("\nCadenas originales recuperadas del disco:")
        for id_f, texto in textos_originales_csv.items():
            print(f"  • ID {id_f}: '{texto}'")

        print("\n" + "="*75)
        print("⚙️ APLICANDO ALGORITMO DE RECONSTRUCCIÓN DINÁMICA...")
        print("="*75)

        # Ejecutamos la función fila por fila inyectando el diccionario de strings recuperados
        df['siniestro_fecha_limpia'] = df.apply(lambda row: reconstruir_fecha_dinamica(row, textos_originales_csv), axis=1)

        print("\n" + "="*75)
        print("✅ VERIFICACIÓN DE DATOS RECONSTRUIDOS (DESPUÉS DE LA CORRECCIÓN)")
        print("="*75)

        # Extraemos los mismos 3 registros ya corregidos para mostrar en la lista de salida
        df_despues = df[df['id_feu'].isin(ids_fallas)].copy()
        df_despues['siniestro_fecha'] = df_despues['siniestro_fecha_limpia']

        # Calculamos variables de calendario para probar que todo es compatible
        df_despues['mes_calculado'] = df_despues['siniestro_fecha'].dt.month
        df_despues['dia_semana_num'] = df_despues['siniestro_fecha'].dt.dayofweek

        nombres_dias = {0: 'Lunes', 1: 'Martes', 2: 'Miércoles', 3: 'Jueves', 4: 'Viernes', 5: 'Sábado', 6: 'Domingo'}
        df_despues['dia_semana_nombre'] = df_despues['dia_semana_num'].map(nombres_dias)
        df_despues['fecha_formateada'] = df_despues['siniestro_fecha'].dt.strftime('%d/%m/%Y')

        columnas_despues = ['id_feu', 'fecha_formateada', 'mes_calculado', 'dia_semana_nombre']
        print(df_despues[columnas_despues].to_string(index=False))
        print("="*75)

        # Escribir los cambios definitivos en el DataFrame global
        df['siniestro_fecha'] = df['siniestro_fecha_limpia']
        df['mes'] = df['siniestro_fecha'].dt.month
        df['dia_semana'] = df['siniestro_fecha'].dt.dayofweek
        df['es_fin_de_semana'] = df['dia_semana'].apply(lambda x: 1 if x >= 5 else 0)

        df.drop(columns=['siniestro_fecha_limpia'], inplace=True, errors='ignore')
        print("\n[OK] El DataFrame 'df' global ha sido actualizado y corregido dinámicamente de forma exitosa.")

    else:
        print("[ERROR] No se detectó un DataFrame llamado 'df' en el entorno de Colab.")
```

### **- Selección de Datos (Justificación de Inclusión/Exclusión)**
*   **Criterios de inclusión:** Se seleccionarán las variables que impactan directamente en la siniestralidad, como la ubicación geográfica, factores temporales (hora, día) y el tipo de vehículo.
*   **Criterios de exclusión:** Se descartarán columnas con un alto porcentaje de datos nulos que no aporten valor predictivo o información administrativa interna del portal que no sea relevante para el modelo.

Aquí están las justificaciones por las que se eligieron cada una de las variables de interés:

1. **Dimensión Temporal (Cuándo ocurre)**
    * *siniestro_fecha y siniestro_hora:* Son las variables con mayor poder predictivo. Permiten al modelo identificar patrones cíclicos:
        * **Estacionalidad:** ¿Hay más choques en vacaciones o durante el ciclo lectivo y jornada laboral?
        * **Ciclo Semanal:** Los siniestros de un viernes noche (ocio/alcohol) son distintos a los de un lunes mañana (hora pico laboral).
        * **Franjas Horarias:** La visibilidad y el volumen de tráfico cambian drásticamente entre el mediodía y la madrugada.

2. **Dimensión Espacial y de Infraestructura (Dónde ocurre)**
    * *latitud, longitud, nombre_via, altura_km:* Estas columnas permiten realizar un análisis de puntos críticos (Blackspots). El modelo puede aprender que ciertas esquinas o tramos de avenidas en Corrientes (como la Av. Independencia o la Av. Maipú) tienen una peligrosidad inherente.

    * *material_de_la_calzada y estado_de_la_calzada:* El agarre del neumático cambia si es asfalto, tierra o si la calzada está mojada o con aceite. Esto ayuda a predecir la probabilidad de éxito de una frenada de emergencia.

3. **Dimensión Ambiental (Condiciones externas)**

4. **Dimensión de la Dinámica del Siniestro (Cómo ocurre)**

    * *ilesos, heridos, fallecidos, cantidad_de_involucrados:*

```python
from geopy.geocoders import Nominatim, ArcGIS
from geopy.extra.rate_limiter import RateLimiter
import numpy as np
import pandas as pd

# Límites del bounding box de Corrientes Capital
LAT_MIN, LAT_MAX = -27.6, -27.4
LON_MIN, LON_MAX = -58.9, -58.6

# Inicialización de dos geocodificadores
geolocator_nom = Nominatim(user_agent="tesis_vial_corrientes_v2_2026")
geolocator_arc = ArcGIS()

geocode_nom = RateLimiter(geolocator_nom.geocode, min_delay_seconds=1.5, error_wait_seconds=5.0)
geocode_arc = RateLimiter(geolocator_arc.geocode, min_delay_seconds=0.5)

# Bounding box para forzar resultados dentro de Corrientes en Nominatim
VIEWBOX = [(LON_MIN, LAT_MAX), (LON_MAX, LAT_MIN)]  # (oeste-norte, este-sur)

def obtener_coordenadas_con_fallback(row):
    """
    Intenta geocodificar con Nominatim (con viewbox).
    Si falla o cae fuera del perímetro, reintenta con ArcGIS.
    Devuelve (lat, lon, metodo).
    """
    calle = row['nombre_via']
    if calle == "SIN_DATOS":
        return np.nan, np.nan, "Sin Datos (Calle Vacía)"

    altura = ""
    if pd.notna(row.get('altura_km')) and str(row.get('altura_km')).strip() not in ['NaN', '']:
        altura = str(row['altura_km']).split('.')[0]

    # Construimos la query con y sin número para mayor resiliencia
    queries = [
        f"{calle} {altura}, Corrientes Capital, Argentina".strip(),
        f"{calle}, Corrientes Capital, Argentina",
    ]

    def dentro_del_perimetro(lat, lon):
        return (LAT_MIN <= lat <= LAT_MAX) and (LON_MIN <= lon <= LON_MAX)

    # --- Intento 1: Nominatim con viewbox ---
    for query in queries:
        try:
            loc = geocode_nom(query, viewbox=VIEWBOX, bounded=True)
            if loc and dentro_del_perimetro(loc.latitude, loc.longitude):
                return loc.latitude, loc.longitude, "Nominatim (viewbox)"
        except Exception:
            pass

    # --- Intento 2: ArcGIS como fallback ---
    for query in queries:
        try:
            loc = geocode_arc(query)
            if loc and dentro_del_perimetro(loc.latitude, loc.longitude):
                return loc.latitude, loc.longitude, "ArcGIS Fallback"
        except Exception:
            pass

    return np.nan, np.nan, "API Falló (ambos)"


def evaluar_contencion_mejorada(row):
    """
    Orquesta la geocodificación con fallback jerárquico:
    1. API (Nominatim + ArcGIS)
    2. Centroide de la calle (calculado con datos históricos válidos)
    3. Centroide por intersección (entre_calle_1 + nombre_via)  ← NUEVO
    4. Centroide global de la ciudad (último recurso, etiquetado explícitamente)
    """
    calle = row['nombre_via']
    lat, lon, metodo = obtener_coordenadas_con_fallback(row)

    if pd.notna(lat):
        return pd.Series([lat, lon, metodo])

    # Fallback 2: Centroide por calle principal
    if calle in centroides_calles.index and pd.notna(centroides_calles.loc[calle, 'latitud']):
        return pd.Series([
            centroides_calles.loc[calle, 'latitud'],
            centroides_calles.loc[calle, 'longitud'],
            "Centroide Calle"
        ])

    # Fallback 3 (NUEVO): Cruce de calles — buscar intersección
    calle2 = row.get('entre_calle_1', 'SIN_DATOS')
    if calle2 != 'SIN_DATOS' and calle2 in centroides_calles.index:
        return pd.Series([
            centroides_calles.loc[calle2, 'latitud'],
            centroides_calles.loc[calle2, 'longitud'],
            "Centroide Calle Transversal"
        ])

    # Fallback 4: Centro de la ciudad (último recurso — MARCAR claramente)
    return pd.Series([centroide_global_lat, centroide_global_lon, "⚠️ Centroide Global (Sin Info)"])
```

```python
import numpy as np
import pandas as pd
import time
import re
from geopy.geocoders import Nominatim, ArcGIS
from geopy.extra.rate_limiter import RateLimiter

# =================================================================
# CONFIGURACIÓN GLOBAL (definir una sola vez)
# =================================================================

TAMANO_LOTE = 100
LAT_MIN, LAT_MAX = -27.6, -27.4
LON_MIN, LON_MAX = -58.9, -58.6
VIEWBOX = [(LON_MIN, LAT_MAX), (LON_MAX, LAT_MIN)]

geolocator_nom = Nominatim(user_agent="tesis_vial_corrientes_v2_2026", timeout=10)
geolocator_arc = ArcGIS()
geocode_nom = RateLimiter(
    geolocator_nom.geocode,
    min_delay_seconds=1.5,
    error_wait_seconds=10.0,  # espera más antes de reintentar tras un error
    max_retries=3,            # intenta 3 veces antes de rendirse
    swallow_exceptions=True   # no interrumpe el loop si falla, pasa al siguiente registro
)
geocode_arc = RateLimiter(geolocator_arc.geocode, min_delay_seconds=0.5)





# =================================================================
# FUNCIONES DE GEOCODIFICACIÓN (definir una sola vez)
# =================================================================

def recalcular_centroides(df):
    """Recalcula los centroides a partir de coordenadas válidas actuales."""
    coords_validas = df[
        df['latitud'].notna() & df['longitud'].notna() &
        df['latitud'].between(LAT_MIN, LAT_MAX) &
        df['longitud'].between(LON_MIN, LON_MAX)
    ]
    centroides = coords_validas.groupby('nombre_via')[['latitud', 'longitud']].median()
    lat_global = coords_validas['latitud'].median()
    lon_global = coords_validas['longitud'].median()
    return centroides, lat_global, lon_global


def dentro_del_perimetro(lat, lon):
    return (LAT_MIN <= lat <= LAT_MAX) and (LON_MIN <= lon <= LON_MAX)


def obtener_coordenadas_con_fallback(row):
    calle = row['nombre_via']
    if calle == "SIN_DATOS":
        return np.nan, np.nan, "Sin Datos (Calle Vacía)"

    altura = ""
    if pd.notna(row.get('altura_km')) and str(row.get('altura_km')).strip() not in ['NaN', '']:
        altura = str(row['altura_km']).split('.')[0]

    queries = [
        f"{calle} {altura}, Corrientes Capital, Argentina".strip(),
        f"{calle}, Corrientes Capital, Argentina",
    ]

    for query in queries:
        try:
            loc = geocode_nom(query, viewbox=VIEWBOX, bounded=True)
            if loc and dentro_del_perimetro(loc.latitude, loc.longitude):
                return loc.latitude, loc.longitude, "Nominatim (viewbox)"
        except Exception:
            pass

    for query in queries:
        try:
            loc = geocode_arc(query)
            if loc and dentro_del_perimetro(loc.latitude, loc.longitude):
                return loc.latitude, loc.longitude, "ArcGIS Fallback"
        except Exception:
            pass

    return np.nan, np.nan, "API Falló (ambos)"


def evaluar_contencion(row, centroides_calles, centroide_global_lat, centroide_global_lon):
    calle = row['nombre_via']
    lat, lon, metodo = obtener_coordenadas_con_fallback(row)

    if pd.notna(lat):
        return pd.Series([lat, lon, metodo])

    if calle in centroides_calles.index and pd.notna(centroides_calles.loc[calle, 'latitud']):
        return pd.Series([centroides_calles.loc[calle, 'latitud'],
                          centroides_calles.loc[calle, 'longitud'],
                          "Centroide Calle"])

    calle2 = row.get('entre_calle_1', 'SIN_DATOS')
    if calle2 != 'SIN_DATOS' and calle2 in centroides_calles.index:
        return pd.Series([centroides_calles.loc[calle2, 'latitud'],
                          centroides_calles.loc[calle2, 'longitud'],
                          "Centroide Calle Transversal"])

    return pd.Series([centroide_global_lat, centroide_global_lon, "⚠️ Centroide Global (Sin Info)"])


# =================================================================
# PIPELINE PRINCIPAL — ejecutar esta celda para procesar TODOS los lotes
# =================================================================

def procesar_geocodificacion_completa(df_filtrado, tamano_lote=TAMANO_LOTE, pausa_entre_lotes=5):
    """
    Procesa todos los registros pendientes de geocodificación en lotes.
    Puede interrumpirse con Ctrl+C y retomarse sin perder progreso,
    ya que guarda los resultados de forma persistente en df_filtrado.

    Parámetros:
        df_filtrado      : DataFrame principal de trabajo
        tamano_lote      : Registros por lote (default 100)
        pausa_entre_lotes: Segundos de espera entre lotes para no saturar la API
    """

    # Inicializar columnas si no existen todavía
    if 'lat_final' not in df_filtrado.columns:
        df_filtrado['lat_final'] = np.nan
        df_filtrado['lon_final'] = np.nan
        df_filtrado['metodo_origen'] = pd.Series(dtype='object')

    numero_lote = 1

    while True:
        centroides_calles, lat_global, lon_global = recalcular_centroides(df_filtrado)

        filas_pendientes = df_filtrado[
            df_filtrado['lat_final'].isna() &
            (df_filtrado['nombre_via'] != 'SIN_DATOS') &
            df_filtrado['latitud'].isna()
        ]

        total_restante = len(filas_pendientes)

        if total_restante == 0:
            print("\n🎉 ¡Geocodificación completa! El 100% de los registros fue procesado.")
            break

        print(f"\n{'='*70}")
        print(f"📦 LOTE {numero_lote} — Pendientes: {total_restante} registros")
        print(f"{'='*70}")

        lote_actual = filas_pendientes.head(tamano_lote).copy()

        # Paso A: Consulta a APIs
        print("🛰️  Consultando APIs de geocodificación...")
        lote_actual[['lat_geopy', 'lon_geopy']] = lote_actual.apply(
            lambda r: pd.Series(obtener_coordenadas_con_fallback(r)[:2]), axis=1
        )
        exitos = lote_actual['lat_geopy'].notna().sum()
        print(f"   API devolvió {exitos}/{len(lote_actual)} coordenadas válidas.")

        # Paso B: Contención perimetral + fallback jerárquico
        lote_actual[['lat_final', 'lon_final', 'metodo_origen']] = lote_actual.apply(
            lambda r: evaluar_contencion(r, centroides_calles, lat_global, lon_global), axis=1
        )

        # Paso C: Guardar persistentemente en df_filtrado
        df_filtrado.loc[lote_actual.index, 'lat_final'] = lote_actual['lat_final']
        df_filtrado.loc[lote_actual.index, 'lon_final'] = lote_actual['lon_final']
        df_filtrado.loc[lote_actual.index, 'metodo_origen'] = lote_actual['metodo_origen']

        # -----------------------------------------------------------------
        # 📋 REPORTE 1: Query enviada vs respuesta cruda de la API
        # -----------------------------------------------------------------
        lote_actual['query_enviada'] = lote_actual.apply(
            lambda r: f"{r['nombre_via']} {r['altura_km'] if pd.notna(r['altura_km']) else ''}, Corrientes, Arg".strip(),
            axis=1
        )
        print(f"\n{'='*120}")
        print("📋 REPORTE 1 — TEXTO ENVIADO VS RESULTADO CRUDO DE LA API")
        print(f"{'='*120}")
        display(lote_actual[['id_feu', 'query_enviada', 'latitud', 'lat_geopy', 'lon_geopy']].head(20))

        # -----------------------------------------------------------------
        # 📋 REPORTE 2: Resultado final con método aplicado
        # -----------------------------------------------------------------
        print(f"\n{'='*120}")
        print("📋 REPORTE 2 — CONTROL DE CONTENCIÓN GEOGRÁFICA Y MÉTODO FINAL")
        print(f"{'='*120}")
        display(lote_actual[['id_feu', 'nombre_via', 'altura_km', 'lat_geopy', 'lat_final', 'lon_final', 'metodo_origen']].head(20))

        # Resumen cuantitativo del lote
        conteo = lote_actual['metodo_origen'].value_counts()
        print("\n📈 Efectividad de este lote:")
        for metodo, cantidad in conteo.items():
            print(f"   • {metodo}: {cantidad} registros")

        restantes_ahora = total_restante - len(lote_actual)
        print(f"\n✅ Lote {numero_lote} guardado. Quedan {max(0, restantes_ahora)} registros.")

        numero_lote += 1

        if restantes_ahora > 0:
            print(f"⏳ Esperando {pausa_entre_lotes}s antes del próximo lote...")
            time.sleep(pausa_entre_lotes)

df_filtrado.to_csv("drive/MyDrive/Proyecto Final de Carrera (2026)/df_filtrado_checkpoint.csv", index=False)

# --- Llamada al pipeline ---
procesar_geocodificacion_completa(df_filtrado)


# =================================================================
# RESUMEN FINAL DE CALIDAD (ejecutar al terminar)
# =================================================================

print("\n📊 RESUMEN FINAL DE MÉTODOS DE GEOCODIFICACIÓN:")
print(df_filtrado['metodo_origen'].value_counts(normalize=False).to_string())
print("\n📊 PORCENTAJES:")
print(df_filtrado['metodo_origen'].value_counts(normalize=True).mul(100).round(1).astype(str).add('%').to_string())
```

# **Imputación y normalización de variables para MACHINE LEARNING:**



```python
import pandas as pd
import numpy as np

print("🧼 Iniciando el saneamiento preventivo de caracteres corrompidos...")

# Creamos una copia de df_filtrado ANTES de imputar
df_saneado = df_filtrado.copy()

# ==============================================================================
# 1. DICCIONARIO CORREGIDO CON VARIANTES DE ENCODING
# ==============================================================================
# Mapeamos combinaciones exactas de bytes UTF-8 en minúsculas y mayúsculas
mapa_limpieza = {
    # Casos críticos específicos observados en tu reporte
    'DÃ­a': 'DIA', 'DÃA': 'DIA', 'DÃ A': 'DIA', 'DÃ a': 'DIA',
    'IntersecciÃ³n': 'INTERSECCION', 'INTERSECCIÃ“N': 'INTERSECCION',
    'SemÃ¡foro': 'SEMAFORO', 'SEMÃFORO': 'SEMAFORO',
    'ColisiÃ³n': 'COLISION', 'COLISIÃ“N': 'COLISION',
    'VehÃ­culos': 'VEHICULOS', 'VEHÃCULOS': 'VEHICULOS',
    'ExplosiÃ³n': 'EXPLOSION', 'EXPLOSIÃ“N': 'EXPLOSION',
    'CaÃ­da': 'CAIDA', 'CAÃDA': 'CAIDA', 'Angulo': 'ANGULO',

    # Reemplazos individuales de caracteres rotos
    'Ã­': 'I', 'Ã': 'I',
    'Ã³': 'O', 'Ã“': 'O',
    'Ã¡': 'A', 'Ã': 'A',
    'Ãº': 'U', 'Ãš': 'U',
    'Ã©': 'E', 'Ã‰': 'E',
    'Ã±': 'N', 'Ã‘': 'N',

    # Acentos estándar
    'Á': 'A', 'É': 'E', 'Í': 'I', 'Ó': 'O', 'Ú': 'U', 'Ñ': 'N',
    '´': '', '`': '', '"': '', "'": ''
}

# ==============================================================================
# 2. FUNCIÓN DE REPARACIÓN (Primero limpia el Mojibake, luego pasa a MAYÚSCULAS)
# ==============================================================================
def normalizar_valores_celda(valor):
    if pd.isna(valor):
        return valor

    # Lo tomamos como string crudo, respetando las minúsculas/mayúsculas de origen
    texto = str(valor).strip()

    # Primero reemplazamos las cadenas rotas exactas
    for termino_errado, termo_correcto in mapa_limpieza.items():
        texto = texto.replace(termino_errado, termo_correcto)

    # Ahora que está limpio, lo pasamos de forma segura a MAYÚSCULAS estándares
    texto = texto.upper()

    # Una segunda pasada de limpieza por si quedó algún acento residual suelto
    for termino_errado, termo_correcto in mapa_limpieza.items():
        if len(termino_errado) == 1: # Solo letras individuales
            texto = texto.replace(termino_errado.upper(), termo_correcto)

    return texto

# ==============================================================================
# 3. APLICACIÓN MASIVA A LAS COLUMNAS DE TEXTO
# ==============================================================================
colunas_texto = df_saneado.select_dtypes(include=['object']).columns

for col in colunas_texto:
    df_saneado[col] = df_saneado[col].apply(normalizar_valores_celda)

print("   [OK] Diccionario aplicado. Caracteres UTF-8 restaurados y estandarizados.")

# ==============================================================================
# 4. CONTROL DE CALIDAD — VERIFICACIÓN DE VARIABLES SANEADAS
# ==============================================================================
print("\n" + "="*60)
print("✨ CONTROL DE CALIDAD: STRINGS SANEADAS")
print("="*60)
print(f"• Luminosidad:             -> {df_saneado['luminosidad'].dropna().unique()[:3]}")
print(f"• Configuración de la Via: -> {df_saneado['configuracion_de_la_via'].dropna().unique()[:3]}")
print(f"• Semáforo:                -> {df_saneado['semaforo'].dropna().unique()[:3]}")
print(f"• Tipo de Siniestro:       -> {df_saneado['tipo_siniestro_multiple'].dropna().unique()[:2]}")
print("==============================================================")

# Guardamos el resultado de vuelta sobre df_filtrado
df_filtrado = df_saneado.copy()
print("🚀 Dataset df_filtrado actualizado en limpio. Ya podés ejecutar la celda 3.5.")
```

```python
print("--- DICCIONARIO DE VARIABLES ---")
print(df_filtrado.info())
print("\nDataset listo para la fase de Modelado.")

# =================================================================
# 3.5. IMPUTACIÓN ESTRATÉGICA DE VARIABLES PARA MACHINE LEARNING
# =================================================================

print("Iniciando imputación estratégica de variables para Machine Learning...")

df_imputado = df_filtrado.copy()

# Columnas que excluimos explícitamente
columnas_excluidas = [
    'latitud', 'longitud', 'lat_final', 'lon_final', 'metodo_origen',
    'id_feu', 'archivo_origen', 'siniestro_fecha', 'siniestro_hora',
    'id_origen_feu'
]

for col in df_imputado.columns:
    if col in columnas_excluidas:
        continue

    nulos_en_columna = df_imputado[col].isna().sum()
    if nulos_en_columna == 0:
        continue

    if pd.api.types.is_numeric_dtype(df_imputado[col]):
        mediana_val = df_imputado[col].median()
        if pd.isna(mediana_val):
            mediana_val = 0
        df_imputado[col] = df_imputado[col].fillna(mediana_val)
        print(f"   🔹 [{col}] ({nulos_en_columna} nulos) -> Imputado con Mediana: {mediana_val}")

    elif pd.api.types.is_object_dtype(df_imputado[col]) or isinstance(df_imputado[col].dtype, pd.CategoricalDtype):
        porcentaje_nulos = nulos_en_columna / len(df_imputado)

        if porcentaje_nulos > 0.50:
            df_imputado[col] = df_imputado[col].fillna('NO_APLICA')
            print(f"   🔸 [{col}] ({nulos_en_columna} nulos - {porcentaje_nulos:.1%}) -> Falta crítica: 'NO_APLICA'")
        else:
            moda_series = df_imputado[col].mode()
            moda_val = moda_series[0] if not moda_series.empty else 'SIN_DATOS'
            df_imputado[col] = df_imputado[col].fillna(moda_val)
            print(f"   ✨ [{col}] ({nulos_en_columna} nulos) -> Imputado con Moda: '{moda_val}'")

# ==============================================================================
# CONTROL FINAL DE CONSISTENCIA
# ==============================================================================
print("\n" + "="*60)
print("✅ PIPELINE AUTOMÁTICO FINALIZADO")
print("="*60)

columnas_control = [c for c in df_imputado.columns if c not in columnas_excluidas]
nulos_restantes = df_imputado[columnas_control].isna().sum().sum()

print(f"• Total de registros en el dataset: {len(df_imputado)}")
print(f"• Nulos remanentes para el modelo: {nulos_restantes}")

if nulos_restantes > 0:
    detalle = df_imputado[columnas_control].isna().sum()
    print("\n⚠️  Columnas con nulos remanentes:")
    print(detalle[detalle > 0].to_string())

print("=" * 60)

# =================================================================
# 3.6. FORMATEO Y DICCIONARIO (Final Dataset)
# =================================================================

df_final = df_imputado.reset_index(drop=True)

print("--- DICCIONARIO DE VARIABLES (Resumen) ---")
print(df_final.info())
print("\nDataset listo para la fase de Modelado.")
```

# **4. Modelado (Modeling)**

## **Los 3 pasos obligatorios antes de Random Forest y XGBoost**

### **Paso 1:** Ingeniería de Características (Feature Engineering)

Hay datos valiosos escondidos que los árboles de decisión van a aprovechar muchísimo mejor si se los simplifica:
*   **Agrupación de categorías:** En variables como "nombre_via" se tiene decenas de nombres únicos. Si se hace One-Hot Encoding directo, van a crearse 300 columnas y van a sufrir la "maldición de la dimensionalidad". Conviene agrupar las calles en categorías macro: Avenida, Calle, Ruta (por ejemplo, detectando con un .str.contains()).
*   **Variables cíclicas:** El "mes", "dia_semana" y "hora" son numéricos, pero para el modelo el mes 12 (Diciembre) está lejos del mes 1 (Enero), cuando en la vida real están pegados. Se puede dejarlos como categóricos o numéricos limpios para el árbol, pero asegurándose de no meter ruido.

### **Paso 2:** Codificación de Variables Categóricas (Encoding)

Este es el puente real hacia el modelo. Tener que convertir todo el texto a números.
*   **One-Hot Encoding (pd.get_dummies):** Ideal para variables categóricas sin un orden jerárquico (ej: material_de_la_calzada $\rightarrow$ Asfalto, Tierra, Hormigón; o luminosidad $\rightarrow$ Día, Noche).
*   **Eliminar columnas administrativas:** Variables como id_feu, siniestro_fecha y siniestro_hora deben salir del set de entrenamiento. No aportan poder predictivo generalizable (el ID es secuencial y las fechas exactas del pasado no se repiten en el futuro).

### **Paso 3:** División del Dataset (Train/Test Split) y Balanceo

1.   Hay que separar la variable objetivo target_gravedad ($Y$) de las características predictivas ($X$).
2.   Dividir en un 70% para entrenar los modelos y un 30% para testearlos y evaluar qué tan bien aprendieron.
3.   **¡OJO!** La siniestralidad vial suele tener un problema grave de clases desbalanceadas (hay muchísimos más siniestros con heridos leves o ilesos que con fallecidos). Si se entrenan los modelos directamente con ese desbalance, el algoritmo se va a volver "perezoso" y va a predecir siempre la clase mayoritaria. Se va a necesitar aplicar técnicas como SMOTE o ajustar el parámetro de pesos de clase (class_weight) dentro de los modelos.

## 3 decisiones inteligentes para el desarrollo de la tesis:

1.  **Clasificación de Vías:** En lugar de crear 300 columnas para cada calle, el script lee nombre_via y clasifica el siniestro de forma macro en: Avenida, Ruta o Calle. Esto le da un patrón estructural clave al modelo.
2.  **Exclusión de Basura Administrativa:** Eliminamos de la matriz de entrenamiento id_feu, siniestro_fecha, siniestro_hora, latitud y longitud (las originales con nulos), asegurándonos de que uses lat_final y lon_final que ya están 100% limpias.
3.  **One-Hot Encoding Inteligente:** Convertimos todas las columnas de texto a variables Dummy (0 y 1) de manera masiva y automática.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

print("⚙️ Iniciando Pipeline Avanzado (Ingeniería de Características + Filtro de Horas Pico)...")

# 1. Usamos el dataset normalizado y limpio de caracteres raros
df_ml = df_final.copy()

# ==============================================================================
# PASO 1: INGENIERÍA DE CARACTERÍSTICAS VIALES Y TEMPORALES
# ==============================================================================

# 📌 NUEVO: Extracción analítica del AÑO antes de eliminar la fecha completa
# Usamos errors='coerce' y dt.year para asegurar que extraiga un número entero limpio
df_ml['anio'] = pd.to_datetime(df_ml['siniestro_fecha'], errors='coerce').dt.year

# Si por algún motivo quedara algún nulo en el año, lo completamos con la moda (el año más frecuente)
if df_ml['anio'].isna().sum() > 0:
    df_ml['anio'] = df_ml['anio'].fillna(df_ml['anio'].mode()[0])

df_ml['anio'] = df_ml['anio'].astype(int)

print(f"   [OK] Variable temporal macro 'anio' extraída correctamente.")


# A. Clasificación macro de las arterias
def clasificar_tipo_via(calle):
    calle_upper = str(calle).upper()
    if 'AV' in calle_upper or 'AVENIDA' in calle_upper:
        return 'Avenida'
    elif 'RUTA' in calle_upper or 'RN' in calle_upper or 'RP' in calle_upper:
        return 'Ruta'
    else:
        return 'Calle_Interna'

df_ml['tipo_via_macro'] = df_ml['nombre_via'].apply(clasificar_tipo_via)

# B. Clasificación de FRANJAS HORARIAS (Dinámica cultural local)
def asignar_franja_horaria(h):
    if 0 <= h < 6:
        return 'Madrugada'
    elif 6 <= h < 13:
        return 'Manana'
    elif 13 <= h < 16:
        return 'Siesta'      # Factor cultural de baja densidad/velocidad
    elif 16 <= h < 21:
        return 'Tarde'
    else:
        return 'Noche'

df_ml['franja_horaria'] = df_ml['hora'].apply(asignar_franja_horaria)

# C. Identificación de HORA PICO (Ingresos, egresos y comercio en Corrientes)
def es_hora_pico(h):
    # Rangos estimativos de alta densidad: 07-09 (Escuelas/Administración), 12-13 (Salida), 17-19 (Comercio)
    if (7 <= h <= 9) or (12 <= h <= 13) or (17 <= h <= 19):
        return 1
    return 0

df_ml['hora_pico'] = df_ml['hora'].apply(es_hora_pico)

print("   [OK] Variables temporales creadas con éxito: 'franja_horaria' y 'hora_pico'.")

# ==============================================================================
# PASO 2: SELECCIÓN DE VARIABLES Y EXCLUSIÓN
# ==============================================================================
y = df_ml['target_gravedad']
print(f"Distribución del target:\n{y.value_counts()}")

# Agregamos 'siniestro_hora' a la lista para eliminar, ya que extrajimos todo su valor en 'hora'
columnas_para_dropear = [
    'id_feu', 'siniestro_fecha', 'siniestro_hora', 'target_gravedad',
    'latitud', 'longitud',
    'nombre_via', 'altura_km', 'entre_calle_1', 'entre_calle_2',
    'heridos', 'ilesos', 'fallecidos', 'cantidad_de_involucrados',
    'metodo_origen'
]

X_crudo = df_ml.drop(columns=[col for col in columnas_para_dropear if col in df_ml.columns])

# ==============================================================================
# PASO 3: ONE-HOT ENCODING AUTOMÁTICO
# ==============================================================================
columnas_categoricas = X_crudo.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"» Aplicando One-Hot Encoding (incluyendo Franjas Horarias)...")
X = pd.get_dummies(X_crudo, columns=columnas_categoricas, drop_first=True)

# Convertimos booleanos a enteros (0/1) para XGBoost
columnas_bool = X.select_dtypes(include=['bool']).columns
X[columnas_bool] = X[columnas_bool].astype(int)

# --- Eliminar solo varianza cero (nada más) ---
selector = VarianceThreshold(threshold=0.0)
mascara  = selector.fit(X).get_support()
X        = X.loc[:, mascara]

# ==============================================================================
# PASO 4: DIVISIÓN DEL DATASET (TRAIN / TEST SPLIT)
# ==============================================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=y
)


# Cálculo dinámico de filas para el reporte
filas_train = X_train.shape[0]
filas_test = X_test.shape[0]
anios_detectados = f"{df_ml['anio'].min()} - {df_ml['anio'].max()}"

print("⚙️ INICIANDO PIPELINE AVANZADO: INGENIERÍA DE CARACTERÍSTICAS Y PREPARACIÓN DE MATRICES")
print("==========================================================================================")
print("[ETAPA 1] EXTRACCIÓN Y CONSTRUCCIÓN DE VARIABLES DE CONTEXTO LOCAL")
print("------------------------------------------------------------------------------------------")
print(f"   细 [Temporal Macro] Extrayendo 'anio' desde 'siniestro_fecha'...")
print(f"      -> Conteo de registros procesados: {len(df_ml)}")
print(f"      -> Rango temporal detectado: [{anios_detectados}]")
print(f"      -> Estado de nulos: 0% (Completado con éxito).")
print("\n   🛣️ [Geográfica Urbana] Creando variable 'tipo_via_macro'...")
print("      -> Aplicando heurística basada en nomenclatura vial de Corrientes Capital.")
print("\n   ⏱️ [Temporal Micro] Evaluando dinámicas culturales de Corrientes...")
print("      -> Variable 'franja_horaria': Segmentación en [Madrugada, Manana, Siesta, Tarde, Noche].")
print("         *Nota metodológica: Se resalta el impacto del factor 'Siesta' por su particularidad local.*")
print("      -> Variable 'hora_pico': Identificación de rangos de alta congestión (07-09, 12-13, 17-19).")

print("\n[ETAPA 2] SANITATION Y SELECCIÓN DE VECTORES DE ENTRADA (FEATURE SELECTION)")
print("------------------------------------------------------------------------------------------")
print("   ✂️ Removiendo variables identificadoras, temporales crudas y geográficas nativas:")
print("      [X] 'id_feu', 'siniestro_fecha', 'siniestro_hora', 'latitud', 'longitud'")
print("      [X] 'nombre_via', 'altura_km', 'entre_calle_1', 'entre_calle_2'")
print("   🎯 Variable Objetivo (Target) aislada: 'target_gravedad'")

print("\n[ETAPA 3] TRATAMIENTO DE CATEGORÍAS (ONE-HOT ENCODING)")
print("------------------------------------------------------------------------------------------")
print(f"   » Aplicando One-Hot Encoding masivo (Dummy variables con drop_first=True)...")
print("   ⚙️ Convirtiendo estructuras booleanas resultantes a formato entero (0/1) para XGBoost.")

print("\n[ETAPA 4] PARTICIÓN DEL DATASET (TRAIN / TEST SPLIT)")
print("------------------------------------------------------------------------------------------")
print("   ⚖️ Aplicando división aleatoria balanceada (Stratified Split por 'target_gravedad'):")
print(f"      -> Proporción fijada: 70% Entrenamiento ({filas_train} filas) / 30% Testeo ({filas_test} filas).")
print("      -> Semilla de aleatoriedad (Random State): 42 (Garantiza replicabilidad).")

print(f"\nX_train: {X_train.shape} | X_test: {X_test.shape}")
print(f"Clase 0 train: {(y_train==0).sum()} | Clase 1 train: {(y_train==1).sum()}")

print("\n" + "="*90)
print("🎯 PIPELINE FINALIZADO: MATRICES CON CONTEXTO HORARIO Y ESPACIAL CONSOLIDADAS")
print("="*90)
print(f"   • Registros Totales en el Experimento:    {len(df_ml)} filas")
print(f"   • Matriz de Entrenamiento (X_train):      {filas_train} filas  |  {X_train.shape[1]} columnas (Features)")
print(f"   • Matriz de Testeo de Control (X_test):   {filas_test} filas  |  {X_test.shape[1]} columnas (Features)")
print(f"   • Vector de Etiquetas (y_train / y_test): Listo para clasificación supervisada.")
print("\n🧬 Atributos clave integrados numéricamente:")
print("   [✓] 'anio' (Tendencia interanual macro)")
print("   [✓] 'hora_pico' (Binaria de alta densidad)")
print("   [✓] 'franja_horaria_SIESTA' (Indicador sociocultural incorporado)")
print("\nPipeline cerrado con 0 nulos detectados. Los datos están listos para los modelos.")
print("==========================================================================================")
```

# **Código de Entrenamiento y Evaluación: Random Forest**

Esta celda va a entrenar el modelo, calcular la matriz de confusión, las curvas de evaluación y guardará las métricas clave en memoria para la comparación final.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_curve,
    auc,
    precision_recall_curve
)

print("🌲 Entrenando clasificador Random Forest con contexto horario y espacial...")

# 1. Instanciamos el modelo con hiperparámetros estables y control de aleatoriedad
rf_model = RandomForestClassifier(
    n_estimators=200,       # Cantidad de árboles
    max_depth=12,           # Evita el sobreajuste (overfitting)
    random_state=42,        # Replicabilidad científica
    class_weight='balanced' # Compensación por desbalanceo de gravedad en Corrientes
)

# 2. Entrenamiento del algoritmo
rf_model.fit(X_train, y_train)
print("   [OK] Modelo entrenado correctamente.")

# 3. Predicciones en el set de testeo (datos no vistos)
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)[:, 1] # Probabilidades para curvas ROC/PR

# ==============================================================================
# MÉTRICAS ACADÉMICAS EN TEXTO
# ==============================================================================
print("\n" + "="*60)
print("📊 REPORTE DE CLASIFICACIÓN DETALLADO (RANDOM FOREST)")
print("="*60)
print(classification_report(y_test, y_pred_rf, target_names=['Leve/Moderado', 'Grave/Fatal']))
print("==============================================================")

# Guardamos la matriz de confusión en una variable
cm_rf = confusion_matrix(y_test, y_pred_rf)

# ==============================================================================
# BLOQUE GRÁFICO: EVALUACIÓN MULTI-CRITERIO
# ==============================================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.set_theme(style="white")

# Gráfico A: Matriz de Confusión Estilizada
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', ax=axes[0], cbar=False,
            xticklabels=['Leve/Mod', 'Grave/Fatal'], yticklabels=['Leve/Mod', 'Grave/Fatal'])
axes[0].set_title('Matriz de Confusión (Random Forest)', fontsize=12, fontweight='bold', pad=10)
axes[0].set_xlabel('Predicción del Modelo', fontsize=11)
axes[0].set_ylabel('Realidad (Acta Policial)', fontsize=11)

# Gráfico B: Curva ROC (Capacidad de discriminación)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

axes[1].plot(fpr_rf, tpr_rf, color='darkorange', lw=2, label=f'Curva ROC (AUC = {roc_auc_rf:.3f})')
axes[1].plot([0, 1], [0, 1], color='navy', lw=1.5, linestyle='--')
axes[1].set_xlim([0.0, 1.0])
axes[1].set_ylim([0.0, 1.05])
axes[1].set_xlabel('Tasa de Falsos Positivos (1 - Especificidad)', fontsize=11)
axes[1].set_ylabel('Tasa de Verdaderos Positivos (Sensibilidad)', fontsize=11)
axes[1].set_title('Curva ROC - Rendimiento Global', fontsize=12, fontweight='bold', pad=10)
axes[1].legend(loc="lower right")
axes[1].grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()

# ==============================================================================
# VARIABLE CLAVE: IMPORTANCIA DE LAS CARACTERÍSTICAS (FEATURE IMPORTANCE)
# ==============================================================================
importancias = rf_model.feature_importances_
indices = np.argsort(importancias)[::-1]

# Armamos un DataFrame con el top 10 para tu tesis
top_features_rf = pd.DataFrame({
    'Variable': X_train.columns[indices[:10]],
    'Importancia': importancias[indices[:10]]
})

plt.figure(figsize=(10, 4))
sns.barplot(data=top_features_rf, x='Importancia', y='Variable', palette='viridis', hue='Variable', legend=False)
plt.title('Top 10 Variables más Predictivas en Siniestralidad (Random Forest)', fontsize=13, fontweight='bold', pad=12)
plt.xlabel('Poder Predictivo Relativo', fontsize=11)
plt.ylabel('Feature', fontsize=11)
plt.tight_layout()
plt.show()
```

## **3. Gráfico de Calibración (Calibration Curve / Reliability Diagram)**

Crucial para validar si las probabilidades que tira el Random Forest reflejan el riesgo real de la calle. Si el modelo dice que un grupo de esquinas tiene 0.7 (70%) de probabilidad de sufrir un siniestro grave, en la realidad histórica el 70% de esos accidentes debieron haber sido graves. Si la línea se aleja mucho de la diagonal de 45°, significa que el modelo necesita calibración.

### **¿Qué estamos viendo en el gráfico? (Explicación para el PFC)**

En la teoría de Machine Learning, un modelo perfectamente calibrado es aquel cuyas predicciones se alinean exactamente sobre la diagonal gris de 45° (Calibración Perfecta). Si tu modelo agrupa un conjunto de esquinas de Corrientes y les asigna un $0.6$ (60%) de probabilidad de sufrir un siniestro grave, lo ideal es que en la realidad histórica el 60% de esos casos sean efectivamente graves.

**1. Tendencia a la Sobreestimación (Curva por encima de la diagonal)**

En casi todo el recorrido (desde probabilidades de $0.35$ hasta $0.75$), la línea naranja se posiciona por encima de la línea punteada ideal.
*   **Interpretación técnica:** Cuando la curva real está por encima de la diagonal en un gráfico donde el eje Y es la Proporción Real, significa que el modelo es conservador en sus estimaciones probabilísticas. Por ejemplo, para el punto donde el modelo predice una probabilidad media de $0.45$, la proporción real de casos graves es de casi $0.60$.
*   **Qué significa en la calle:** El algoritmo tiende a asignar probabilidades numéricas más bajas de lo que el peligro real representa. Esto suele ocurrir de manera intencional cuando entrenamos modelos usando el parámetro class_weight='balanced', ya que penalizamos fuertemente al algoritmo si se le "escapa" un caso grave. El modelo prefiere marcar el riesgo, aunque numéricamente tire una probabilidad moderada.

**2. Subestimación Ligera en Rangos Altos ($0.75$ a $0.90$)**

Fijémonos en el pequeño bache que hace la línea naranja entre el $0.75$ y el $0.85$. Ahí la curva cruza y se posiciona por debajo de la diagonal.
*   En ese tramo específico, cuando el modelo se pone "súper seguro" y asigna probabilidades en torno al $0.85$, la realidad muestra que la tasa de fatalidad es ligeramente menor ($0.77$). Hay una pequeña sobreconfianza en ese umbral superior.

**3. El Ajuste de los Extremos**

Lo más positivo del gráfico es el último punto, cercano al $0.92$. Ahí la línea naranja vuelve a tocar de manera casi perfecta la diagonal ideal. Esto demuestra que cuando el Random Forest detecta configuraciones urbanas combinadas de altísimo riesgo, su predicción probabilística es casi matemáticamente exacta.

```python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import make_scorer, roc_auc_score
from xgboost import XGBClassifier

print("🎯 Buscando mejores hiperparámetros para XGBoost (versión corregida)...")

param_grid = {
    'n_estimators':     [200, 300, 500],
    'max_depth':        [3, 4, 5, 6],
    'learning_rate':    [0.01, 0.05, 0.1],
    'subsample':        [0.7, 0.8, 0.9],
    'colsample_bytree': [0.6, 0.7, 0.8],
    'min_child_weight': [1, 3, 5],
    'gamma':            [0, 0.1, 0.2],
}

# Sin scale_pos_weight — usamos class_weight equivalente via sample_weight
xgb_base = XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    tree_method='hist'   # más estable numéricamente en Colab
)

search = RandomizedSearchCV(
    xgb_base,
    param_distributions=param_grid,
    n_iter=50,
    scoring='roc_auc',   # string directo, más estable que make_scorer
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    random_state=42,
    n_jobs=-1,
    verbose=1,
    error_score=0        # en vez de nan, pone 0 si hay error en algún fold
)

# Calculamos sample_weight para compensar el desbalanceo
from sklearn.utils.class_weight import compute_sample_weight
pesos = compute_sample_weight(class_weight='balanced', y=y_train)

search.fit(X_train, y_train, sample_weight=pesos)
xgb_optimizado = search.best_estimator_

print(f"\n✅ Mejor AUC en CV: {search.best_score_:.4f}")
print(f"   Hiperparámetros encontrados:")
for k, v in search.best_params_.items():
    print(f"   {k}: {v}")
```

```python
import pandas as pd
import re
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter


df_coordenadas = df.copy()
# 1. Aislamos las primeras 20 filas que no tienen coordenadas originales
df_prueba = df_coordenadas[df['latitud'].isnull()].head(20).copy()

# Definimos el geolocalizador
geolocator = Nominatim(user_agent="tesis_vial_corrientes_2026")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1.2)


def buscar_coords(row):
    if pd.isna(row['latitud']) and row['nombre_via'] != "SIN_DATOS":
        altura = str(row['altura_km']) if (pd.notna(row['altura_km']) and str(row['altura_km']).strip() != 'NaN') else ""
        # Armamos la query con la calle ya normalizada
        query = f"{row['nombre_via']} {altura}, Corrientes, Argentina"
        try:
            location = geocode(query)
            if location:
                return pd.Series([location.latitude, location.longitude])
        except:
            return pd.Series([None, None])
    return pd.Series([row['latitud'], row['longitud']])


# FUNCIÓN DE NORMALIZACIÓN (Tu lógica impecable)
def normalizar_calle(calle):
    if pd.isna(calle) or str(calle).strip().upper() in ['S/D', '0', 'NAN', 'NONE', '']:
        return "SIN_DATOS"

    # Convertimos a string y aseguramos mayúsculas
    c = str(calle).upper().strip()

    # ==============================================================================
    # 0. REPARACIÓN DE MOJIBAKE (Errores de codificación de acentos y Ñ)
    # ==============================================================================
    mojibake_correcciones = {
        'Ã': 'E', 'Ã': 'A', 'Ã': 'I', 'Ã': 'O', 'Ã': 'U', # Acentos rotos en mayúsculas
        'Ã¡': 'A', 'Ã©': 'E', 'Ã­': 'I', 'Ã³': 'O', 'Ãº': 'U', # Acentos rotos en minúsculas
        'Ã±': 'N', 'Ã': 'N',                                   # La letra Ñ / ñ
        'Â': '', '': ''                                        # Caracteres huérfanos comunes
    }
    for roto, limpio in mojibake_correcciones.items():
        c = c.replace(roto, limpio)

    # ==============================================================================
    # 1. CORRECCIÓN DE ERRORES TIPOGRÁFICOS ESPECÍFICOS (DICCIONARIO DE ERRATAS)
    # ==============================================================================
    erratas_comunes = {
        r'\bRATU\b': 'RUTA', r'\bAVNEDIA\b|\bAVENDIA\b': 'AV',
        r'\bINDEPNDENCIA\b|\bINDEP.\b|\bINDEP\b|\bINDEPENCIA\b': 'INDEPENDENCIA',
        r'\bCHACABUCOCO\b': 'CHACABUCO', r'\bINDEPENDENCIA\b': 'AV INDEPENDENCIA',
        r'\bGUTEMBERG\b': 'GUTENBERG',
        r'\bSARMIENO\b': 'SARMIENTO', r'\bS. MARTIN\b|\bSANMARTIN\b': 'SAN MARTIN',
        r'\bGRAL. PAZ\b': 'GENERAL PAZ'
    }
    for patron, correccion in erratas_comunes.items():
        c = re.sub(patron, correccion, c)

    # 2. Reemplazar vocales con acentos remanentes por vocales limpias
    c = c.replace('Á', 'A').replace('É', 'E').replace('Í', 'I').replace('Ó', 'O').replace('Ú', 'U')

    # 3. Estandarizar abreviaturas viales
    c = re.sub(r'\bAVENIDA\b|\bAV\.\b|\bAV_\b', 'AV', c)
    c = re.sub(r'\bAV AV\b', 'AV', c)
    c = re.sub(r'^(CALLE\s+)', '', c)
    c = re.sub(r'\bRUTA NACIONAL\b|\bRTANAC\b|\bRN\b', 'RUTA', c)

    # 4. Limpiar caracteres especiales de conflicto y dobles espacios internos
    c = re.sub(r'[^\w\s]', ' ', c)
    c = re.sub(r'\s+', ' ', c).strip()
    return c

df_prueba['nombre_via'] = df_prueba['nombre_via'].apply(normalizar_calle)
df_prueba['entre_calle_1'] = df_prueba['entre_calle_1'].apply(normalizar_calle)
print("   [OK] Nombres de calles normalizados en la muestra.")


print("Iniciando geocodificación de prueba (aproximadamente 25 segundos)...")

# Guardamos el resultado en las nuevas columnas de prueba
df_prueba[['lat_geopy', 'lon_geopy']] = df_prueba.apply(buscar_coords, axis=1)

exitos = df_prueba['lat_geopy'].notnull().sum()
print(f"🎉 ¡Prueba terminada! De 20 intentos, se localizaron con éxito: {exitos} esquinas.\n")

# =================================================================
# COMPONENTE DE REVISIÓN VISUAL (Lo que necesitabas agregar)
# =================================================================
print("="*120)
print("📋 REPORTE COMPARATIVO: TEXTO ENVIADO VS RESULTADO GEOPY")
print("="*120)

# Creamos una columna temporal en el reporte para ver qué texto exacto se armó
df_reporte = df_prueba.copy()
df_reporte['query_enviada'] = df_reporte.apply(
    lambda r: f"{r['nombre_via']} {r['altura_km'] if pd.notna(r['altura_km']) else ''}, Corrientes, Arg", axis=1
)

columnas_visibles = ['id_feu', 'query_enviada', 'latitud', 'lat_geopy', 'lon_geopy']
print(df_reporte[columnas_visibles].to_string(index=False))
print("="*120)


import numpy as np
import pandas as pd


# 1. Definimos los límites perimetrales estrictos de Corrientes Capital
LAT_MIN, LAT_MAX = -27.6, -27.4
LON_MIN, LON_MAX = -58.9, -58.6


# 2. Primero calculamos los centroides limpios del dataset original para usarlos de auxilio
centroides_calles = df.groupby('nombre_via')[['latitud', 'longitud']].median()
centroide_global_lat = df['latitud'].median()
centroide_global_lon = df['longitud'].median()


def buscar_coords_seguro(row):
    calle = row['nombre_via']
    altura = str(row['altura_km']) if (pd.notna(row['altura_km']) and str(row['altura_km']).strip() != 'NaN') else ""

    # Si ya tiene coordenadas originales dentro de rango, las dejamos intactas
    if pd.notna(row['latitud']) and (LAT_MIN <= row['latitud'] <= LAT_MAX) and (LON_MIN <= row['longitud'] <= LON_MAX):
        return pd.Series([row['latitud'], row['longitud'], "Original"])


    # Si no tiene (o estaban rotas), disparamos GeoPy
    if calle != "SIN_DATOS":
        query = f"{calle} {altura}, Corrientes, Argentina"
        try:
            location = geocode(query)
            if location:
                lat, lon = location.latitude, location.longitude


                # === FILTRO DE SEGURIDAD CRÍTICO ===
                # Validamos si la coordenada de GeoPy pertenece REALMENTE a Corrientes Capital
                if (LAT_MIN <= lat <= LAT_MAX) and (LON_MIN <= lon <= LON_MAX):
                    return pd.Series([lat, lon, "GeoPy Exitoso"])
                else:
                    # Falso positivo detectado (Se fue a Goya, Libres, etc.) -> Aplicamos Centroide Calle
                    if calle in centroides_calles.index and pd.notna(centroides_calles.loc[calle, 'latitud']):
                        return pd.Series([centroides_calles.loc[calle, 'latitud'], centroides_calles.loc[calle, 'longitud'], "Centroide Calle (Falso Positivo)"])
        except:
            pass # Si falla la API, cae al plan de respaldo de abajo

    # Plan de respaldo final: Centroide de la calle o Global si la calle no tiene antecedentes GPS
    if calle in centroides_calles.index and pd.notna(centroides_calles.loc[calle, 'latitud']):
        return pd.Series([centroides_calles.loc[calle, 'latitud'], centroides_calles.loc[calle, 'longitud'], "Centroide Calle (API Falló)"])

    return pd.Series([centroide_global_lat, centroide_global_lon, "Centroide Global"])


# 3. Ejecutamos la corrección inteligente sobre tu dataframe de prueba
print("Iniciando geocodificación con auditoría perimetral...")
df_prueba[['lat_final', 'lon_final', 'metodo_origen']] = df_prueba.apply(buscar_coords_seguro, axis=1)


# =================================================================
# IMPRESIÓN DEL REPORTE DE CONTROL REPARADO
# =================================================================
print("\n" + "="*120)
print("📋 REPORTE REPARADO: CONTROL DE CONTENCIÓN GEOGRÁFICA")
print("="*120)
columnas_reporte = ['id_feu', 'nombre_via', 'altura_km', 'lat_geopy', 'lat_final', 'lon_final', 'metodo_origen']
print(df_prueba[columnas_reporte].to_string(index=False))
print("="*120)

```

# **Marco Legal y Normativo:**

El marco legal y normativo de datos abiertos en la ciudad de Corrientes (Capital) se fundamenta en una serie de ordenanzas municipales y normativas nacionales que promueven la transparencia, el gobierno abierto y la accesibilidad de la información pública.

## Marco Normativo Local (Corrientes Capital)
*   **Ordenanza Nº 6885 (2020):** Es la norma fundamental que crea el marco para la apertura de datos en el ámbito de la Municipalidad de la Ciudad de Corrientes. Establece la implementación progresiva de la apertura de datos para fortalecer la transparencia y la rendición de cuentas.
*   **Portal de Datos Abiertos (datos.ciudaddecorrientes.gov.ar):** Creado como una herramienta de Gobierno Abierto, aloja bases de datos municipales en formatos reutilizables (CSV, JSON, XML).
*   **Carta Orgánica Municipal:** Garantiza la transparencia de la gestión municipal.
*   **Dirección General de Innovación Pública y Gobierno Abierto:** Área encargada de articular con las distintas dependencias municipales para la publicación de los datos.

## Marco Normativo Provincial y Nacional
*   **Ley Provincial N° 5.834:** Garantiza el derecho de acceso a la información pública en la Provincia de Corrientes.
*   **Ley Nacional N° 25.326 de Protección de Datos Personales (Habeas Data):** Limita la publicación de datos abiertos, prohibiendo la difusión de datos personales, sensibles o que afecten la privacidad de los ciudadanos.
*   **Ley Nacional N° 27.275 de Acceso a la Información Pública (2016):** Marco general que promueve la transparencia activa y la publicación de datos en formatos abiertos.

**Características de la Apertura de Datos**
*   **Principios:** Los datos deben ser libres, accesibles, gratuitos, no propietarios y reutilizables.
*   **Temáticas:** El portal municipal incluye datos de movilidad, salud pública, seguridad, economía, ambiente y desarrollo urbano.
*   **Protección:** La apertura busca un equilibrio entre la transparencia y la protección de datos personales tanto de funcionarios como de ciudadanos.

## **Columnas de los dataset:**

1. **Identificación y Control (Metadados)**

    * *id_feu:* corresponde al identificador único o número de referencia del Formulario Estadístico Único (FEU)

2. **Temporalidad (Cuándo ocurrió)**

    * *nombre_via / altura_km:* El nombre de la calle y la numeración catastral o el kilómetro.
    * *entre_calle_1 / entre_calle_2:* Calles transversales; sirven para identificar intersecciones (esquinas).

    * *peatones:* Cantidad de peatones involucrados en el evento.

5. **Características del Siniestro**

    * *tipo_colision:* Detalle técnico del impacto (Frontal, Lateral, Por alcance, Ángulo).
    * *despiste_previo:* Indica si el vehículo perdió el control de la trayectoria antes del impacto.

6. **Entorno e Infraestructura**
    * *trazado_via:* Geometría de la calle (Recta, Curva, Intersección, Encrucijada).

    * *material_de_la_calzada:* Tipo de superficie (Asfalto, Hormigón, Tierra, Ripio).
    * *estado_de_la_calzada:* Condición del suelo (Seca, Húmeda, Con aceite, Con arena).
    * *division_fisica_de_la_via / barreras_de_seguridad:* Presencia de elementos de protección o separación.
    * *senalizacion_carteleria / horizontal:* Indica si había señales verticales (carteles) o marcas en el asfalto.
    * *semaforo:* Indica si hay semáforo y si estaba funcionando, intermitente o apagado.

7. **Condiciones Ambientales**
    * *luminosidad:* Si era de día, noche con luz artificial, o noche sin iluminación.
    * *luz_artificial:* Calidad de la iluminación pública en el momento del choque.
    * *estado_ambiental:* Clima al momento del siniestro (Despejado, Lluvia, Niebla, Nublado).
    * *visibilidad:* Si la visión del conductor estaba despejada o afectada por factores externos.

```python
import pandas as pd
import glob
import numpy as np

# Listamos todos los archivos CSV subidos, especificando la carpeta
archivos = glob.glob('drive/MyDrive/Proyecto Final de Carrera (2026)/Archivos Siniestros 2018-2025/*.csv')

dfs = []
for archivo in archivos:
    # Leemos cada archivo (ajustamos el separador si es necesario)
    df_temp = pd.read_csv(archivo, low_memory=False)
    # Agregamos una columna para saber de qué archivo vino (útil para auditoría)
    df_temp['archivo_origen'] = archivo
    dfs.append(df_temp)

# Unimos todos los años en un solo gran DataFrame
# 'sort=False' mantiene las columnas aunque no coincidan perfectamente entre años
df = pd.concat(dfs, axis=0, ignore_index=True, sort=False)

print(f"Total de registros cargados: {df.shape[0]}")








# 1. Ver una muestra aleatoria de 10 filas
print("Muestra aleatoria de los datos combinados:")
display(df.sample(10))

# 2. Verificar cuántos registros hay por cada año (archivo de origen)
print("\nConteo de registros por archivo de origen:")
print(df['archivo_origen'].value_counts())

# 3. Ver información general de las columnas (tipos de datos y nulos)
print("\nResumen de estructura y valores nulos:")
df.info()
```

# **2)- Limpieza y Normalización**

Aquí están las justificaciones por las que se eligieron las variables de interés:

1. **Dimensión Temporal (Cuándo ocurre)**
    * *siniestro_fecha y siniestro_hora:* Son las variables con mayor poder predictivo. Permiten al modelo identificar patrones cíclicos:
        * **Estacionalidad:** ¿Hay más choques en vacaciones o durante el ciclo lectivo y jornada laboral?
        * **Ciclo Semanal:** Los siniestros de un viernes noche (ocio/alcohol) son distintos a los de un lunes mañana (hora pico laboral).
        * **Franjas Horarias:** La visibilidad y el volumen de tráfico cambian drásticamente entre el mediodía y la madrugada.

2. **Dimensión Espacial y de Infraestructura (Dónde ocurre)**
    * *latitud, longitud, nombre_via, altura_km:* Estas columnas permiten realizar un análisis de puntos críticos (Blackspots). El modelo puede aprender que ciertas esquinas o tramos de avenidas en Corrientes (como la Av. Independencia o la Av. Maipú) tienen una peligrosidad inherente.
    * *via_publica y zona_ocurrencia:* Diferencian el comportamiento del conductor en una calle barrial frente a una ruta o avenida, donde las velocidades son mayores y los tipos de impacto cambian.
    * *material_de_la_calzada y estado_de_la_calzada:* El agarre del neumático cambia si es asfalto, tierra o si la calzada está mojada o con aceite. Esto ayuda a predecir la probabilidad de éxito de una frenada de emergencia.
    * *semaforo:* Es una variable de control. Su presencia o ausencia (o mal funcionamiento) es un predictor clave en colisiones en intersecciones.

3. **Dimensión Ambiental (Condiciones externas)**
    * *estado_ambiental y estado_fisico_ambiental:* En Corrientes, las lluvias intensas reducen la visibilidad y aumentan la distancia de frenado. Un modelo que ignore el clima perdería mucha precisión.
    * *luminosidad:* Crucial para predecir siniestros con peatones o motos, que son menos visibles durante la noche o en zonas con mala iluminación artificial (dia/noche).
    * *luz_artificial:* Indica en caso de ser de noche la variable luminosidad sus valores pueden ser SI/NO.

4. **Dimensión de la Dinámica del Siniestro (Cómo ocurre)**
    * *tipo_siniestro_multiple, unico, tipo_colision, despiste_previo:* Estas columnas describen la naturaleza del evento.
        * Ejemplo: Un "despiste previo" suele estar asociado a exceso de velocidad o fatiga, mientras que una "colisión lateral" suele estar ligada a fallos en el respeto de prioridades de paso.
    * *vehiculos y peatones:* Definen la complejidad del evento y ayudan a clasificar la configuración del choque (auto-auto, auto-moto, auto-peatón).

5. Dimensión de Gravedad (Qué consecuencias tiene)
    * *ilesos, heridos, fallecidos, cantidad_de_involucrados:*
        * Como Variable Objetivo (Y): Estas son las columnas que se usarán para que el modelo aprenda a clasificar si un siniestro será leve o grave.
        * Sin estas columnas, podría predecirse dónde habrá un choque, pero no qué tan grave será.



```python
# 1. Selección de columnas útiles
# Nos quedamos solo con lo que realmente sirve para predecir accidentes
cols_interes = [
    'siniestro_fecha', 'siniestro_hora', 'zona_ocurrencia', 'via_publica',
    'nombre_via', 'altura_km', 'latitud', 'longitud', 'ilesos', 'heridos',
    'fallecidos', 'vehiculos', 'peatones', 'tipo_colision', 'material_de_la_calzada',
    'estado_de_la_calzada', 'estado_fisico_ambiental', 'luminosidad', 'luz_artificial',
    'estado_ambiental', 'semaforo', 'tipo_siniestro_múltiple',
    'tipo_siniestro_unico', 'despiste_previo', 'cantidad_de_involucrados',
]

# Creamos un nuevo dataframe filtrado
df_filtrado = df[cols_interes].copy()

# 2. Limpieza de Coordenadas (Lat/Long)
# Corrientes está aprox en Lat -27 y Long -58.
# Convertimos a número y limpiamos signos de puntuación extraños.
def limpiar_coordenada(valor):
    if pd.isna(valor) or valor == 'S/D': return np.nan
    try:
        # Quitamos puntos extra y dejamos solo el primero si hay varios
        s = str(valor).replace(',', '.')
        partes = s.split('.')
        if len(partes) > 2:
            s = partes[0] + '.' + "".join(partes[1:])
        return float(s)
    except:
        return np.nan

df_filtrado['latitud'] = df_filtrado['latitud'].apply(limpiar_coordenada)
df_filtrado['longitud'] = df_filtrado['longitud'].apply(limpiar_coordenada)

# 3. Formatear Fechas y crear variables temporales
df_filtrado['siniestro_fecha'] = pd.to_datetime(df_filtrado['siniestro_fecha'], errors='coerce')
df_filtrado['mes'] = df_filtrado['siniestro_fecha'].dt.month
df_filtrado['dia_semana'] = df_filtrado['siniestro_fecha'].dt.dayofweek # 0=Lunes, 6=Domingo
df_filtrado['hora_neta'] = pd.to_datetime(df_filtrado['siniestro_hora'], errors='coerce').dt.hour

# 4. Unificar texto (Todo a mayúsculas para evitar 'Seca' vs 'SECA')
cols_texto = ['via_publica', 'estado_de_la_calzada', 'estado_ambiental', 'luminosidad', 'semaforo']
for col in cols_texto:
    df_filtrado[col] = df_filtrado[col].astype(str).str.upper().str.strip()

# 5. Variable Objetivo: Gravedad del Siniestro
# 0: Solo daños, 1: Heridos, 2: Fallecidos
df_filtrado['gravedad'] = 0
df_filtrado.loc[df_filtrado['heridos'] > 0, 'gravedad'] = 1
df_filtrado.loc[df_filtrado['fallecidos'] > 0, 'gravedad'] = 2

print("¡Limpieza completada!")
df_filtrado.head()
```
